```python
# ==========================================
# SISTEMA DE FARMACÉUTICOS EN PYTHON
# ==========================================

# Lista donde se guardarán los farmacéuticos
farmaceuticos = []

# ==========================================
# FUNCIÓN AGREGAR
# ==========================================
def agregarfarmaceutico():
    print("\n--- AGREGAR FARMACÉUTICO ---")

    nombre = input("Nombre: ")
    especialidad = input("Especialidad: ")
    edad = input("Edad: ")

    farmac = {
        "nombre": nombre,
        "especialidad": especialidad,
        "edad": edad
    }

    farmaceuticos.append(farmac)

    print("Farmacéutico agregado correctamente.")

# ==========================================
# FUNCIÓN LEER
# ==========================================
def leerfarmaceutico():
    print("\n--- LEER FARMACÉUTICO ---")

    if len(farmaceuticos) == 0:
        print("No hay farmacéuticos registrados.")
    else:
        posicion = int(input("Ingrese posición del farmacéutico: "))

        if posicion >= 0 and posicion < len(farmaceuticos):
            print("Nombre:", farmaceuticos[posicion]["nombre"])
            print("Especialidad:", farmaceuticos[posicion]["especialidad"])
            print("Edad:", farmaceuticos[posicion]["edad"])
        else:
            print("Posición no válida.")

# ==========================================
# FUNCIÓN LISTAR
# ==========================================
def listarfarmaceutico():
    print("\n--- LISTA DE FARMACÉUTICOS ---")

    if len(farmaceuticos) == 0:
        print("No hay farmacéuticos registrados.")
    else:
        contador = 0

        for f in farmaceuticos:
            print("\nFarmacéutico", contador)
            print("Nombre:", f["nombre"])
            print("Especialidad:", f["especialidad"])
            print("Edad:", f["edad"])

            contador += 1

# ==========================================
# FUNCIÓN BUSCAR
# ==========================================
def buscarfarmaceutico():
    print("\n--- BUSCAR FARMACÉUTICO ---")

    buscar = input("Ingrese nombre a buscar: ")

    encontrado = False

    for f in farmaceuticos:
        if f["nombre"].lower() == buscar.lower():
            print("\nFarmacéutico encontrado")
            print("Nombre:", f["nombre"])
            print("Especialidad:", f["especialidad"])
            print("Edad:", f["edad"])

            encontrado = True

    if encontrado == False:
        print("Farmacéutico no encontrado.")

# ==========================================
# FUNCIÓN MODIFICAR
# ==========================================
def modificarfarmaceutico():
    print("\n--- MODIFICAR FARMACÉUTICO ---")

    buscar = input("Ingrese nombre del farmacéutico: ")

    for f in farmaceuticos:
        if f["nombre"].lower() == buscar.lower():

            print("Ingrese los nuevos datos")

            f["nombre"] = input("Nuevo nombre: ")
            f["especialidad"] = input("Nueva especialidad: ")
            f["edad"] = input("Nueva edad: ")

            print("Farmacéutico modificado correctamente.")
            return

    print("Farmacéutico no encontrado.")

# ==========================================
# FUNCIÓN BORRAR
# ==========================================
def borrarfarmaceutico():
    print("\n--- BORRAR FARMACÉUTICO ---")

    buscar = input("Ingrese nombre del farmacéutico a borrar: ")

    for f in farmaceuticos:
        if f["nombre"].lower() == buscar.lower():
            farmaceuticos.remove(f)

            print("Farmacéutico eliminado correctamente.")
            return

    print("Farmacéutico no encontrado.")

# ==========================================
# MENÚ PRINCIPAL
# ==========================================
while True:

    print("\n=================================")
    print("   SISTEMA DE FARMACÉUTICOS")
    print("=================================")
    print("1. Agregar farmacéutico")
    print("2. Leer farmacéutico")
    print("3. Listar farmacéuticos")
    print("4. Buscar farmacéutico")
    print("5. Modificar farmacéutico")
    print("6. Borrar farmacéutico")
    print("7. Salir")
    print("=================================")

    opcion = input("Seleccione una opción: ")

    if opcion == "1":
        agregarfarmaceutico()

    elif opcion == "2":
        leerfarmaceutico()

    elif opcion == "3":
        listarfarmaceutico()

    elif opcion == "4":
        buscarfarmaceutico()

    elif opcion == "5":
        modificarfarmaceutico()

    elif opcion == "6":
        borrarfarmaceutico()

    elif opcion == "7":
        print("Saliendo del sistema...")
        break

    else:
        print("Opción inválida.")
```
